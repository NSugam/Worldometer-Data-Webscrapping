```python
# Web Scrapping Corona Virus data from Worldometer Website

from bs4 import BeautifulSoup
import requests

url = BeautifulSoup('https://www.worldometers.info/coronavirus', 'html.parser')
response = requests.get(url)
data = response.text
data = BeautifulSoup(data,'lxml')

final = []
tag = data.table
clear = tag.find_all('tr')
list1 = clear[0:-8]  # delete last 7 unnecessary data
for i in list1:
    final.append (i.get_text())


final_data = []
for i in final:
    my_data = i.split('\n')[2:11]  # range for number of column
    final_data.append(my_data)

del final_data[1:8]  # delete first 7 unnecessary data without removing topics

import csv  # to write in csv file
fields = final_data
with open('data/worldometer.csv','w') as file:
    x = csv.writer (file)
    for i in fields:
        x.writerow(i)

import pandas as pd  # to read the csv file data
df = pd.read_csv('data/worldometer.csv')
df.iloc[0:11]
```
